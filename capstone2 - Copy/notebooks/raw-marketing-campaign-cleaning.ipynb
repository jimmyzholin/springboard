This notebook will be the initial data cleaning for the "Marketing Campaign" dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = "../data/raw/marketing_campaign.csv"
```

```python
df = pd.read_csv(filename, sep='\t')
df.head()
```

```python
df.describe()
```

```python
dfColumns = df.columns
```

```python
#Find where null values exist.
df.isnull().sum()
```

This dataset has 2240 records. There are only 24 instances where null values are found and they all appear in the "Income" column. Given the small number of null values relative to the whole dataset, the rows with null values will be removed.

```python
df[df['Income'].isnull()]
```

```python
noNullDf = df[~df['Income'].isnull()]
noNullDf.to_csv("../data/processed/noNullMarketingData.csv")
```
